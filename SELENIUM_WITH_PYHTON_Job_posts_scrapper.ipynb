```python
## build function ##

# GET ALL JOB LINKS 

''' !!!ADVANCED: talk a little about how to find with parcial calss name!!!

e.g.:
      driver.find_elements_by_css_selector(<tagname>"[<attribute>^='<partial text of the attribute>']")

exammple in the line bellow '''

job_raw = driver.find_elements_by_css_selector("a[class^='disabled ember-view']")





def scrapper(numb_j = len(job_raw)+ 2): ## add pages
    
    """ SUMMARY: This function retrives all the job posts 
    links from one page and returns a dataset with
    the name of the job in one column and in the other the link. 
    Also it will write in different files the same info
    for every single job post.
    
    HOW IT WORKS: Input the number of jobs you want to scrape. 
    And it will search in the page the elemnts by css selector 
    of all the job posts then loop for every single element and retrive the 'href'. 
    Also it will click in every job post and find the job Name.
    This info will be saved in a dictionary that will in the end be converted to 
    a dataset.
    Bellow we will open and create a text file with the name of the Job post 
    and then save inside the link for further details"""

    
    # empty dictionary for saving the job names with the link 
    job_dic = {}
    dict_requirements = {}
    
    lista_javascript = []
    lista_python = []
    lista_c = []
    lista_ruby = []
    lista_net = []
    lista_sql = []
    lista_ang = []
    lista_react = []
    lista_node = []
    lista_java = []
    lista_names = []
    lista_location = []
    lista_employees = []
    lista_link = []
    lista_candidate = []
    lista_anun = []
    lista_job_name = []
    lista_body = []
    
    # reduce the page size in order to be able to find the name of the job in the right session 
    driver.execute_script("document.body.style.zoom='67%'")

    # all jobs in the page
    job_raw = driver.find_elements_by_css_selector("a[class^='disabled ember-view']") 

    # go to the end of the page for all the elements to be loaded
    page = driver.find_element_by_css_selector("a[class^='disabled ember-view']")
    page.send_keys(Keys.END)   
    # go to the top of the page for all the elements to be loaded
    page.send_keys(Keys.CONTROL + Keys.HOME) # combination of the two keys bring you to the top of the element
    
    
    for j_idx in range(numb_j):
        
        # get the link of the job
        try:
            ref = job_raw[j_idx].get_attribute('href') 
            lista_link.append(ref)
            time.sleep(2)
        
        
        # increase the page size
            driver.execute_script("document.body.style.zoom='100%'")
        
        ## lets click in the job post ##
        #driver.find_elements_by_css_selector("a[class^='disabled ember-view']")[j_idx].click()
        
            job_raw[j_idx].click()
            time.sleep(1.4)
        except:
            print('fim')
        
        ## then we reduce the page in order to be able to see the right part of the page and find the element with the name of the job ##
        driver.execute_script("document.body.style.zoom='67%'")
        time.sleep(2.3)
        #Gets the job details
        body = driver.find_element_by_id('job-details').text.lower() # Texto do Body 
        body_cleaned = body.replace('\n',' ').replace(',',' ').replace('.',' ').replace('(',' ').replace('\\',' ').replace('/',' ').replace(')',' ').replace(';',' ').replace('+',' ').replace('js',' ')
        words_list = body_cleaned.split(' ')
        lista_body.append(body_cleaned)

        #Gets how many keywords are in the details.   
        
        

        dict_requirements = {
            'Job position' : lista_job_name,
            'Company' : lista_names,
            'Location': lista_location,
            'Dimension' : lista_employees,  
            'Candidates' : lista_candidate,
            'Post time': lista_anun,      
            'JavaScript' : lista_javascript,
            'Python' : lista_python,
            'C#' : lista_c,
            'Ruby' : lista_ruby,
            '.Net' : lista_net,
            'SQL' : lista_sql,
            'Angular.js' : lista_ang,
            'React.js' : lista_react,
            'Node.js' : lista_node,
            'Java' : lista_java,
            'Links' : lista_link,
            'Body' : lista_body
        }

        if 'python' in  words_list:
            lista_python.append(1)        
        else:
            lista_python.append(0)       
        if 'c#' in words_list:
            lista_c.append(1)  
        else:
            lista_c.append(0)
        if 'ruby' in words_list:      
            lista_ruby.append(1)        
        else:
            lista_ruby.append(0)       
        if 'javascript' in words_list:      
            lista_javascript.append(1)        
        else:
            lista_javascript.append(0) 
        if 'net' in words_list:      
            lista_net.append(1)        
        else:
            lista_net.append(0) 
        if ('sql' or 'mysql' or 'postgresql') in words_list:      
            lista_sql.append(1)        
        else:
            lista_sql.append(0) 
        if 'angular' in words_list:      
            lista_ang.append(1)        
        else:
            lista_ang.append(0) 
        if 'react' in words_list:      
            lista_react.append(1)        
        else:
            lista_react.append(0) 
        if 'node' in words_list:      
            lista_node.append(1)        
        else:
            lista_node.append(0) 
        if 'java' in words_list:      
            lista_java.append(1)        
        else:
            lista_java.append(0) 
                
                
        
        
        
        # get the job name with the .text method
#         job_name = driver.find_element_by_css_selector("h2[class^='jobs-details-top-card__job-title']").text
        try:
            job_name = driver.find_element_by_css_selector("h2[class^='jobs-details-top-card__job-title']").text
            lista_job_name.append(job_name)
            time.sleep(1.7)
        except:
            lista_job_name.append('Check link')
            
            
        try:
            company_name = driver.find_element_by_css_selector("a[class^='jobs-details-top-card__company-url t-black--light t-normal ember-view']").text
            lista_names.append(company_name)
            time.sleep(2.2)
        except:
            lista_names.append('Check link')
        
        #Gets the job location
        try:
            job_location = driver.find_element_by_css_selector("a[class^='jobs-details-top-card__exact-location t-black--light link-without-visited-state']").text
            lista_location.append(job_location)
            time.sleep(1.9)
        except:
            div_loc = driver.find_element_by_css_selector("div[class^='jobs-details-top-card__company-info t-14 t-black--light t-normal mt1']").text
            location = div_loc.split('\n')[-1]
            lista_location.append(location)
        
        
        #Gets the number of employees
        try:
            page = driver.find_element_by_css_selector("div[class^='jobs-box__html-content jobs-description-content__text t-14 t-normal']")
            page.send_keys(Keys.END)
            job_employees = driver.find_element_by_css_selector("span[class^='jobs-company__inline-information']").text
            lista_employees.append(job_employees)
            time.sleep(2.3)
        except:
            lista_employees.append('Check link')
            
        try:
            candidate = driver.find_element_by_css_selector("span[class^='jobs-details-job-summary__text--ellipsis']").text
            lista_candidate.append(candidate)
            time.sleep(2.2)
        except:
            lista_candidate.append('Check link')
        # increase the page size
        
        
        try: 
            p_anun = driver.find_element_by_css_selector("p[class^='t-12 t-black--light mt2 mb0']").text
            anun = p_anun.split('\n')[1]
            lista_anun.append(anun)
        except:
            lista_anun.append('Check link')
        driver.execute_script("document.body.style.zoom='100%'")
        
        

#         with open(list(job_dic.keys())[int(j_idx/2)].replace(" ", "_").replace("/","_") + '_link' + '.txt', 'w') as out_file:

#             parsed_line = ref

#             out_file.write(parsed_line)

        if str((j_idx + 1)/2)[-1] == '0':
            print('Job links got: ' + str(int((j_idx + 1)/2)))
            
        else:
            continue
    
    
    Job_dataframe = pd.DataFrame.from_dict(dict_requirements)
    
    
    
#     driver.close() # closes the driver

    return Job_dataframe

```
